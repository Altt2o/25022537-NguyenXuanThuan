Viết hàm kiểm tra một số có phải số Amstronhg hay không? Hàm trả về kiểu bool, có tham số là một số
nguyên x

```python
def is_armstrong(x: int):
    if x < 0:
        return False
    tong = 0
    c = len(str(x))
    temp = x
    while temp > 0:
        b = temp % 10
        tong += b**c
        temp //= 10
    if tong == x:
        return True
    return False
print(is_armstrong(370))
```

W5A19 VIết hàm kiểm tra một số có phải số kỳ diệu hay không, xử lý trên dữ liệu kiểu số nguyên.

```python
def ky_dieu(n):
    while n > 9:
        tong = 0
        temp = n
        while temp > 0:
            b = temp % 10
            tong += b
            temp //= 10 
        n = tong
    return n == 1
print(ky_dieu(19))
```

W5A23 Viết hàm quadratic_roots_count(a, b, c) trả về số nghiệm thực của phương trình a*x^2 + b*x + c = 0

```python
def quadratic_roots_count(a, b, c):
    delta = b**2 - 4*a*c
    if delta > 0:
        return 2
    elif delta == 0:
        return 1
    else:
        return 0
```

Viết hàm point_in_quadrilateral(quad, p) kiểm tra điểm p có nằm trong hoặc trên biên của tứ giác quad hay
không. Trong đó: quad là một tứ giác (4 đỉnh – 4 điểm), p là một điểm cần kiểm tra. VD: quad = [(0,0), (4,0), (4,3),
(0,3)], p = (2,1)

```python
def point_in_quadrilateral(quad, p):
    px, py = p
    n = len(quad)
    inside = False
    for i in range(n):
        j = (i + 1) % n
        x1, y1 = quad[i]
        x2, y2 = quad[j]
        if (x2 - x1) * (py - y1) == (y2 - y1) * (px - x1):
            if min(x1, x2) <= px <= max(x1, x2) and min(y1, y2) <= py <= max(y1, y2):
                return True
            x_intersect = (x2 - x1) * (py - y1) / (y2 - y1) + x1
            if px < x_intersect:
                inside = not inside
    return inside
```

W5A28 Viết hàm tính góc ∠ABC (độ) từ tọa độ A(x1,y1), B(x2,y2), C(x3,y3).

```python
import math
def goc_ABC(x1,y1,x2,y2,x3,y3):
    AB = ((x1-x2)**2 + (y1-y2)**2)**0.5
    AC = ((x1-x3)**2 + (y1-y3)**2)**0.5
    BC = ((x3-x2)**2 + (y3-y2)**2)**0.5
    a = math.degrees((AB**2 + BC**2 - AC**2)/(2*AB*BC))
    return a
```

W5A33
Viết hàm smart_taxi_fare(car_type, start_hour, km, wait_minutes) tính tổng tiền taxi theo quy tắc giá mở cửa,
giá km, phí chờ, phụ phí ban đêm và giảm giá.

```python
def smart_taxi_fare(car_type, start_hour, km, wait_minutes):
    if car_type == 4:
        gia_mo_cua = 11000 
        gia_km = 14500
    elif car_type == 7:
        gia_mo_cua = 12000
        gia_km = 15500
    else:
        return "Loại xe không hợp lệ! Vui lòng nhập 4 hoặc 7."
    if km <= 1:
        tien_duong = gia_mo_cua
    else:
        tien_duong = gia_mo_cua + (km - 1) * gia_km

    phi_cho = 0
    if wait_minutes > 5:
        phi_cho = (wait_minutes - 5) * 1000
    tong_tien = tien_duong + phi_cho
    if start_hour >= 22 or start_hour <= 5:
        tong_tien += tong_tien * 0.10
    if km > 30:
        tong_tien -= tong_tien * 0.10
    return round(tong_tien)
```

Viết hàm sum_1_to_n(n) trả về tổng các số x từ 1 đến n, nếu x có các chữ số đôi một khác nhau
Nếu n ≤ 0 trả về 0.

```python
def sum_1_to_n(n):
    if n <= 0:
        return 0
    else:
        tong = 0
        for i in range(1,n+1):
            if len(str(i)) == len(set(str(i))):
                tong += i
        return tong
print(sum_1_to_n(100))
```
