W4A6
"[SumPrimesInRange]
Nhập hai số nguyên dương a, b (a ≤ b). Hãy tính tổng các số nguyên tố trong đoạn
[a, b]. Mỗi dòng là một bộ test (a b)."


```python
while True:
    a, b = map(int, input().split())
    if a <= b:
        break
    print("Không thỏa mãn, hãy nhập lại a,b")
tong = 0
for n in range(a,b+1):
    if n < 2:
        continue
    la_so_nguyen_to = True
    for i in range(2,n):
        if n % i == 0:
            la_so_nguyen_to = False
            break
    if la_so_nguyen_to == True:
        tong += n
print(tong)
```

W4A8
ReverseAddPalindrome]
Nhập số nguyên dương n. Lặp lại phép toán n = n + reverse(n) cho đến khi n là số
đối xứng (palindrome). In ra số bước và giá trị palindrome thu được. 


```python
n = int(input())
buoc = 0
while True:
    dao_nguoc = 0
    temp = n
    while temp > 0:
        dao_nguoc = dao_nguoc * 10 + (temp % 10)
        temp //= 10
    if n == dao_nguoc:
        break
    n = n + dao_nguoc
    buoc += 1
print(buoc, n)
```

W4A10
"[LongestCollatzUpToN]
Nhập vào số nguyên dương n. Tìm số x trong [1, n] có độ dài dãy Collatz lớn nhất
và in ra x cùng độ dài dãy tương ứng (x L). Nếu có nhiều số cùng độ dài, chọn số
nhỏ nhất."

```python
n = int(input())
do_dai_lon_nhat = 0
so_co_day_lon_nhat = 0
for x in range(1,n+1):
    so_hien_tai = x
    buoc = 1
    while so_hien_tai > 1:
        if so_hien_tai % 2 == 0:
            so_hien_tai //= 2
        else:
            so_hien_tai = 3 * so_hien_tai + 1
        buoc += 1
    if buoc > do_dai_lon_nhat:
        do_dai_lon_nhat = buoc
        so_co_day_lon_nhat = x
print(so_co_day_lon_nhat, do_dai_lon_nhat)


    
```

W4A14
[GCD]
Cho hai số m và n là hai số nguyên dương nhập từ bàn phím. Viết chương trình
tìm ước chung lớn nhất của hai số m và n. 


```python
while True:
    m, n = map(int, input().split())
    if m > 0 and n > 0:
        break
    print("Hãy nhập lại m, n dương")
m_max = 0
n_max = 0
for a in range(1,m):
    if m % a == 0:
        m_max = a
for b in range(1,n):
    if n % b == 0:
        n_max = b
print(m_max, n_max)
```

W4A25
Nhập vào một dãy số nguyên, ngưng nhập khi người dùng nhập -1.
Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập

```python
n_max = None
n_min = None
while True:
    n = int(input())
    if n == -1:
        break
    if n_max is None:
        n_max = n
        n_min = n
    else:
        if n > n_max:
            n_max = n
        elif n < n_min:
            n_min = n
if n_max is not None:
    print(n_max)
    print(n_min)
else:
    print("Hay nhap so khac -1")

```

Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ
bằng tổng hai số trước đó
Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt

```python
a = int(input())
x = 1
y = 1
z = 1
while x + y < a:
    z = x + y
    x = y
    y = z
print(z)
```

Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu ký tự in hoa, bao nhiêu ký tự in
