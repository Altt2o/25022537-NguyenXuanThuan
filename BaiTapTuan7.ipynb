W7A1
Viết hàm tìm kiếm nhị phân
trên một danh sách đã
được sắp xếp tăng dần.
Trả về chỉ số của phần tử
hoặc -1 nếu không tìm
thấy. mẫu hàm: def
binary_search(arr, left,
right, target):

```python
def binary_search(arr, left, right, target):
    if left > right:
        return -1
    mid = (left + right) // 2
    if arr[mid] == target:
        return mid
    elif arr[mid] > target:
        return binary_search(arr, left, mid - 1, target)
    else:
        return binary_search(arr, mid + 1, right, target)
```

Cho danh sách arr và một
số nguyên x, đếm xem x
xuất hiện bao nhiêu lần
trong danh sách. def
count_occurrences(arr, x) -> int:

```python
def count_occurrences(arr, x) -> int:
    tong = 0
    for i in range(len(arr)):
        if arr[i] == x:
            tong += 1
    return tong
arr = list(map(int,input().split()))
x = int(input())
print(count_occurrences(arr,x))
```

W7A3
Viết hàm thực hiện thuật
toán bubble sort, Selection
Sort để sắp xếp danh sách
theo thứ tự tăng dần. Sắp
xếp nổi bọt (Bubble Sort);
Sắp xếp chọn (Selection
Sort)

```python
def xap_sep(arr):
    a = len(arr)
    for i in range(a):
        for j in range(0,a-1-i):
            if arr[j] > arr[j+1]:
                arr[j],arr[j+1] = arr[j+1],arr[j]
    return arr
n = list(map(int,input().split()))
print(xap_sep(n))
```

W7A4
Viết hàm tìm phần tử có số
lần xuất hiện nhiều nhất
trong danh sách. Nếu có
nhiều phần tử cùng số lần,
trả về phần tử xuất hiện
trước.


```python
def so_lan_xuat_hien(n):
    dem_max = 0
    ban_dau = n[0]
    for i in n:
        dem = n.count(i)
        if dem > dem_max:
            dem_max = dem
            ban_dau = i
        elif dem == dem_max:
            break
    return ban_dau,dem_max
n = list(map(int,input().split()))
so, lan = so_lan_xuat_hien(n)
print(f"{so} xuat hien nhieu nhat, som nhat, {lan} lan")
```

W7A8
Bạn được cung cấp: một
danh sách các khoảng thời
gian interval = [[l1, r1], [l2,
r2], ...] và một danh sách
các truy vấn queries = [q1,
q2, ...]
Với mỗi truy vấn q, hãy tìm
độ dài nhỏ nhất của
khoảng thời gian [li, ri] sao
cho li ≤ q ≤ ri.
Nếu không tồn tại khoảng
thời gian nào như vậy, hãy
trả về -1 cho truy vấn đó.
solve in < O(m.n), hint:
binary search in the
queries list

```python
def solve_w7a8_co_ban(intervals, queries):
    ket_qua = []
    for q in queries:
        do_dai_nho_nhat = -1
        for interval in intervals:
            l = interval[0]
            r = interval[1]
            if l <= q and q <= r:
                do_dai_hien_tai = r - l + 1
                if do_dai_nho_nhat == -1 or do_dai_hien_tai < do_dai_nho_nhat:
                    do_dai_nho_nhat = do_dai_hien_tai
        ket_qua.append(do_dai_nho_nhat)
    return ket_qua
```

W7A13
Dãy hài hòa là một dãy
thỏa mãn điều kiện giá trị
lớn nhất và giá trị nhỏ nhất
của dãy khác nhau đúng
bằng \( 1 \).
Viết hàm trả về độ dài của
dãy hài hòa con dài nhất có
thể có trong mảng giá trị
nguyên \( arr \) có \( n \)
phần tử
Ví dụ:
Input: 1 3 2 2 5 2 3 7
Output: 5 - Dãy hài hòa con
dài nhất là [3,2,2,2,3

```python

nums = [int(x) for x in input().split() if x.isdigit()]
nums.sort()
max_len = 0
n = len(nums)
for i in range(n):
    for j in range(i + 1, n):
        if nums[j] - nums[i] == 1:
            if j - i + 1 > max_len:
                max_len = j - i + 1
        elif nums[j] - nums[i] > 1:
            break
print(max_len)
```
