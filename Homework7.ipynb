Cho một mảng số nguyên không âm. Gọi mảng đó là "mảng đẹp" nếu
mỗi phần tử x trong mảng xuất hiện đúng x lần.
Kiểm tra xem mảng có phải là mảng đẹp hay không. Nếu không hãy xóa
đi y số (y là nhỏ nhất có thể) để mảng còn lại thành mảng đẹp.


```python
n = list(map(int,input().split()))
nhom = []
for i in set(n):
    if n.count(i) > i:
        nhom.append(n.count(i) - i)
    elif n.count(i) < i:
        nhom.append(n.count(i))
print(sum(nhom))
```

W7A19
Merge 2 mảng đã sắp xếp (có thể tăng dần, có thể giảm dần) mà không
dùng bất cứ thuật toán sắp xếp nà

```python
x = list(map(int,input().split()))
y = list(map(int,input().split()))
for i in range(len(x)):
    for j in range(i+1,len(x)):
        if x[i] < x[j]:
            x[i],x[j] = x[j],x[i]
for i in range(len(y)):
    for j in range(i+1,len(y)):
        if y[i] > y[j]:
            y[i],y[j] = y[j],y[i]
print(x + y)
```

W7A22
Cho N tên sinh viên gồm cả họ và tên, tìm xem tên nào hay gặp nhất.
Nếu có nhiều tên cùng số lần gặp thì lấy tên có thứ tự từ điển nhỏ nhất

```python
n = int(input())
danh_sach_ten = []
for i in range(n):
    ten = input()
    danh_sach_ten.append(ten)
ten_pho_bien_nhat = ""
so_lan_max = 0
for ten in danh_sach_ten:
    so_lan_xuat_hien = 0
    for t in danh_sach_ten:
        if t == ten:
            so_lan_xuat_hien += 1
    if so_lan_xuat_hien > so_lan_max:
        so_lan_max = so_lan_xuat_hien
        ten_pho_bien_nhat = ten
    elif so_lan_xuat_hien == so_lan_max:
        if ten < ten_pho_bien_nhat:
            ten_pho_bien_nhat = ten
print(ten_pho_bien_nhat)
```

W7A25
Tìm số có đúng k chữ số lớn nhất có thể trong một số có n chữ số (coi
đầu vào là một số rất lớn - str). Không được hoán đổi vị trí các chữ số


```python
chuoi_so = input()
k = int(input())
so_luong_can_xoa = len(chuoi_so) - k
ket_qua = []
for chu_so in chuoi_so:
    while so_luong_can_xoa > 0 and len(ket_qua) > 0 and ket_qua[-1] < chu_so:
        ket_qua.pop()
        so_luong_can_xoa -= 1
    ket_qua.append(chu_so)
ket_qua = ket_qua[:k]
print(*ket_qua)
```

W7A27
Cho một dãy ngoặc ()(){}{}[][]..., kiểm tra xem dãy ngoặc đã cho có hợp lệ
hay không

```python
chuoi_ngoac = input()
ngan_xep = []
hop_le = True
for ngoac in chuoi_ngoac:
    if ngoac == '(' or ngoac == '[' or ngoac == '{':
        ngan_xep.append(ngoac)
    else:
        if len(ngan_xep) == 0:
            hop_le = False
            break
        ngoac_mo_cuoi = ngan_xep.pop()
        if ngoac == ')' and ngoac_mo_cuoi != '(':
            hop_le = False
            break
        if ngoac == ']' and ngoac_mo_cuoi != '[':
            hop_le = False
            break
        if ngoac == '}' and ngoac_mo_cuoi != '{':
            hop_le = False
            break
if hop_le == True and len(ngan_xep) == 0:
    print("Hợp lệ")
else:
    print("Không hợp lệ")
```

W7A28
Bà
̀i toán Tìm dãy các hợp số dài nhất liên tiếp không chứa số nguyên
nào <= N (N là số nguyên dương cho trước). Nếu có nhiều dãy cùng độ
dài thì lấy dãy nhỏ nhất (gặp đầu tiên từ trái qua)

```python
def la_hop_so(x):
    if x <= 3:
        return False
    for i in range(2, x):
        if x % i == 0:
            return True
    return False
n = int(input())
max_len = 0
start_index = -1
current_len = 0
current_start = -1
for i in range(1, n + 1):
    if la_hop_so(i):
        if current_len == 0:
            current_start = i
        current_len += 1
        if current_len > max_len:
            max_len = current_len
            start_index = current_start
    else:
        current_len = 0
if max_len == 0:
    print("Không có dãy hợp số nào.")
else:
    day_ket_qua = []
    for i in range(start_index, start_index + max_len):
        day_ket_qua.append(str(i))
    print(*day_ket_qua)
```

W7A29
Tìm Dãy con tăng dần liên tục dài nhất của một danh sách nhập vào từ
bàn phím. Lấy dãy đầu tiên nếu có nhiều dãy có số phần tử bằng nhau


```python
chuoi_nhap = input()
cac_chu = chuoi_nhap.split()
nums = []
for chu in cac_chu:
    nums.append(int(chu))
if len(nums) == 0:
    print("Danh sách trống")
else:
    max_len = 1
    start_max = 0
    current_len = 1
    current_start = 0
    for i in range(1, len(nums)):
        if nums[i] > nums[i - 1]:
            current_len += 1
            if current_len > max_len:
                max_len = current_len
                start_max = current_start
        else:
            current_len = 1
            current_start = i
    for i in range(start_max, start_max + max_len):
        print(nums[i], end=" ")
```

W7A30
Tìm Dãy con tăng dần KHÔNG liên tục dài nhất của một danh sách nhập
vào từ bàn phím 

```python
chuoi_nhap = input()
cac_chu = chuoi_nhap.split()
nums = []
for chu in cac_chu:
    nums.append(int(chu))
n = len(nums)
if n == 0:
    print("Danh sách trống")
else:
    F = [1] * n
    Vet = [-1] * n
    for i in range(1, n):
        for j in range(0, i):
            if nums[i] > nums[j] and F[j] + 1 > F[i]:
                F[i] = F[j] + 1
                Vet[i] = j
    max_len = F[0]
    vi_tri_ket_thuc = 0
    for i in range(1, n):
        if F[i] > max_len:
            max_len = F[i]
            vi_tri_ket_thuc = i
    day_con = []
    curr = vi_tri_ket_thuc
    while curr != -1:
        day_con.append(nums[curr])
        curr = Vet[curr]
    day_con.reverse()
    for so in day_con:
        print(so, end=" ")
```
