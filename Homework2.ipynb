Bài 3 : Tính tam giác: Nhâp vào ba số a,b,c bất kì. Kiểm tra xem nó có thể là độ dài ba cạnh hay không, nếu không thì in ra màn hình 'Khong la ba canh cua mot tam giac'. Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra mànhình.

```python
a, b, c = map(float, input().split())
if a + b < c or a + c < b or b + c < a:
    print("Không là ba cạnh của 1 tam giác")
else:
    p = (a+b+c)/2
    z = (p*(p-a)*(p-b)*(p-c))**(1/2)
    if a == b != c or a == c != b or c == b != a:
        print("Là tam giác cân. Diện tích là:", z)
    elif a**2 + b**2 == c**2 or a**2 + c**2 == b**2 or c**2 + b**2 == a**2:
        print("Là tam giác vuông. Diện tích là:", z)
    elif a == b == c:
        print("Là tam giác đều. Diện tích là:", z)
    else:
        print("Là tam giác thường. Diện tích là:", z)
```

Bài 5: Viết chương trình giải phương trình bậc 2 (ax2 + bx + c = 0)

```python
a, b, c = map(float, input().split())
delta = b**2 - 4*a*c
if delta > 0:
    print("Phương trình có 2 nghiệm phân biệt là:")
    print("x1 =", (-b+(delta)**(1/2))/(2*a))
    print("x2 =", (-b-(delta)**(1/2))/(2*a))
elif delta == 0:
    print("Phương trình có nghiệm duy nhất là: x =", end=" ")
    print((-b)/(2*a))
else:
    print("Phương trình vô nghiệm")
```

Bài 10: Nhập vào tâm và bán kính của một đường tròn. Sau đó nhập vào một điểm A(x, y) bất kì và kiểm tra xem nó có thuộc đường tròn hay không?

```python
a, b = map(float, input().split())
R = float(input())
x, y = map(float, input().split())
if (x-a)**2 + (y-b)**2 == R**2:
    print("Điểm A thuộc đường tròn")
else:
    print("Diểm A không thuộc đường tròn")
```
