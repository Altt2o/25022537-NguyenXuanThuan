Nhâ ̣p vào mảng n hàng, m cột số nguyên. In lại mảng ra màn hình như ma ̣
trận hai chiều, định dạng mỗi phần tử 4 vị trí căn phải

W6A17
Nhập vào mảng vuông 2 chiều n hàng, n cột, in ra đường chéo chính và
đường chéo phụ của mảng

```python
n = int(input())
matrix = []
for i in range(n):
    m = list(map(int,input().split()))
    matrix.append(m)
chinh = []
for i in range(n):
    chinh.append(matrix[i][i])
phu = []
for j in range(n):
    phu.append(matrix[j][n-1-j])
print("Duong cheo chinh:", chinh)
print("Duong cheo phu:", phu)
```

Nhâ ̣p vào mảng n hàng, m cột số nguyên và số nguyên k (n, m, k nhập trên ̣
cùng một dòng, cách nhau dấu cách). Tính và in ra tổng các phần tử trên cột
k (hàng k)

```python
n,m,k = map(int,input().split())
matrix = []
tong_1 = 0
tong_2 = 0
for i in range(n):
    x = list(map(int,input().split()))
    matrix.append(x)
for i in range(n):
    tong_1 += matrix[i][k]
for j in range(m):
    tong_2 += matrix[k][j]
print(tong_1 + tong_2 - matrix[k][k])
```

W6A19
Viết hàm đưa vào 1 list số nguyên, tìm và trả về vị trí có giá trị lớn nhất
t

```python
n = list(map(int,input().split()))
def vi_tri_max(n):
    if not n:
        return None
    gia_tri = max(n)
    buoc = []
    for i in range(len(n)):
        if n[i] == gia_tri:
            buoc.append(i+1)
    return buoc
print(vi_tri_max(n))
```

W6A20
Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và
trả về vị trí của phần tử đầu tiên có giá trị k trong list, nếu không có thì trả về
-1


```python
lst = list(map(int,input().split()))
k = int(input())
def abc(lst):
    buoc = 0
    gia_tri_k = False
    for i in range(len(lst)):
        if lst[i] == k:
            buoc = i+1
            gia_tri_k = True
            break
    if gia_tri_k:
        return buoc
    else:
        return -1
print(abc(lst))
```

W6A21
Nhập một dòng gồm các chuỗi (tách bởi dấu cách). Tạo dictionary ánh xạ
chuỗi -> tuple các vị trí (0-based) mà chuỗi đó xuất hi

```python
n = input().split()
def anh_xa(n):
    nhom = {}
    for i,j in enumerate(n):
        if j not in nhom:
            nhom[j] = [i]
        else:
            nhom[j].append(i)
    for j in nhom:
        nhom[j] = tuple(nhom[j])
    return nhom
print(anh_xa(n))
```

W6A22
Viết hàm đưa vào 1 list số nguyên, tìm và trả về vị trí có giá trị lớn nhất
trong list

```python
def gia_tri_lon_nhat(n):
    gia_tri_max = max(n)
    buoc = ()
    for i in range(len(n)):
        if n[i] == gia_tri_max:
            buoc += (i,)
    return buoc
n = list(map(int,input().split()))
print(gia_tri_lon_nhat(n))
```

W6A23 VIết hàm xóa đi các số <x có trong mảng


```python
def xoa(n,x):
    nhom = []
    for i in n:
        if i >= x:
            if i not in nhom:
                nhom.append(i)
    return nhom
n = list(map(int,input().split()))
x = int(input())
print(xoa(n,x))
```

W6A24
Viêết hàm đẩy tất cả các số 0 về đầu (về cuối), các số khác vẫn đúng vị trí
trước sau: 0 2 3 2 4 0 0 0 9 => 0 0 0 2 3 2 4 9


```python
def dao(n):
    nhom_1 = []
    nhom_2 = []
    for i in n:
        if i == 0:
            nhom_1.append(i)
        else:
            nhom_2.append(i)
    return nhom_1 + nhom_2
n = list(map(int,input().split()))
print(dao(n))
```

W6A27
Cho một dãy các số nguyên, kiểm tra xem nó là dãy tăng thực sự hay giảm
thực sự, hay vô danh (không tăng, không giảm)

```python
n = list(map(int,input().split()))
tang = False
giam = False
bang = False
for i in range(1,len(n)):
    if n[i] > n[i-1]:
        tang = True
    elif n[i] < n[i-1]:
        giam = True
    else:
        bang = True
if tang and giam == False and bang == False:
    print('Tang')
elif giam and tang == False and bang == False:
    print('Giam')
else:
    print('Vo danh')
```
